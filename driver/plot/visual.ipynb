```python
import  numpy as np
import matplotlib.pyplot as plt
micro_list = np.load('plot/micro.npy') *100
macro_list = np.load('plot/macro.npy') *100
x = range(len(micro_list))
plt.figure(figsize=(30,15))   
plt.plot(x,micro_list, color='r',label='micro')
plt.plot(x,macro_list, color='g',label='macro')
plt.xlabel("epoch")#横坐标名字
plt.ylabel("value")#纵坐标名字
plt.legend(loc = "best")#图例
```
